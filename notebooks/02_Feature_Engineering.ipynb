````python
df["years_since_last_promotion"].describe()
df["performance_rating"].describe()
````

````python
# Create Promotion Due Feature

df["promotion_due"] = np.where(
    (df["years_since_last_promotion"] >= 5) &
    (df["performance_rating"] >= 3),
    "Yes",
    "No"
)
````

🧠 Why did we use np.where() instead of np.select()?

Because this is a binary decision.

Two outcomes:
✅ Yes
✅ No

For binary conditions, np.where() is simpler and more readable.

Situation	NumPy Function
2 outcomes (Yes/No)	np.where() ✅
3 or more outcomes	np.select() ✅



# 🚀 Feature 5 – Income Per Year of Experience

This is one of my favorite features because **it doesn't exist in the original dataset**, yet it provides valuable business insight.

## 🏢 Business Problem

Imagine the HR Director asks:

> **"Are employees with similar experience being paid fairly?"**

Suppose we have two employees:

| Employee | Monthly Income | Total Working Years |
| -------- | -------------: | ------------------: |
| A        |         12,000 |                  12 |
| B        |         12,000 |                   3 |

Both earn ₹12,000.

But Employee B earns much more **relative to experience**.

---

# 🎯 Business Logic

We create a new metric:

```text
Income Per Year of Experience
=
Monthly Income
────────────────────────────
Total Working Years
```

Example:

| Income | Experience | Result |
| -----: | ---------: | -----: |
|  12000 |         12 |   1000 |
|  12000 |          3 |   4000 |

Employee B has a much higher income per year of experience.

---

# ⚠️ Important

Some employees have:

```text
total_working_years = 0
```

If we divide by 0:

```python
12000 / 0
```

Python throws an error (or produces infinity).

So we need to handle that safely.








Example:

| monthly_income | total_working_years | income_per_year_experience |
| -------------: | ------------------: | -------------------------: |
|           5993 |                   8 |                     749.12 |
|           5130 |                  10 |                     513.00 |
|           2090 |                   7 |                     298.57 |



# 🎤 Interview Answer

> "I engineered an `income_per_year_experience` feature to measure salary relative to experience. This provides a normalized metric that can help identify employees who may be overpaid or underpaid compared to peers with similar experience. I used `np.where()` to safely handle employees with zero years of experience and avoid division-by-zero errors."

---

# ⭐ Why this feature is impressive

This isn't just another category like `age_group`.

It's a **derived business metric**, which demonstrates analytical thinking. It can support:

* Compensation analysis

# 💻 Step 1 – Create the Feature

We'll use **`np.where()`** to avoid division by zero.


````python

df["income_per_year_experience"] = np.where(
    df["total_working_years"] == 0,
    df["monthly_income"],
    df["monthly_income"] / df["total_working_years"]
)
````

# 💻 Step 2 – Round the Values


````python
df["income_per_year_experience"] = df["income_per_year_experience"].round(2)
````

# 💻 Step 3 – Verify


````python

df[[
    "monthly_income",
    "total_working_years",
    "income_per_year_experience"
]].head(10)
````

# 🚀 Feature 6 – Employee Tenure Category

## 🏢 Business Problem

The HR Manager asks:

> **"How many new employees, experienced employees, and long-term employees do we have?"**

The dataset has:

```text
years_at_company
```

Example:

| years_at_company |
| ---------------: |
|                1 |
|                5 |
|               12 |
|               25 |

These are numbers.

For business reporting, it's easier to analyze categories.

---

# 🎯 Business Rule

| Years at Company | Tenure Category      |
| ---------------: | -------------------- |
|              0–2 | New Employee         |
|              3–5 | Established Employee |
|             6–10 | Experienced Employee |
|              11+ | Veteran Employee     |

This helps HR answer questions like:

* Which tenure group has the highest attrition?
* Are new employees leaving too early?
* Do veteran employees receive more promotions?

# 🎤 Interview Answer

> "I created an `employee_tenure_category` feature to segment employees based on their tenure with the organization. This helps HR analyze retention, promotion trends, and attrition across different tenure groups instead of relying only on raw numerical values."




````python


# 💻 Step 1 – Create Conditions

tenure_conditions = [
    (df["years_at_company"] <= 2),
    (df["years_at_company"] >= 3) & (df["years_at_company"] <= 5),
    (df["years_at_company"] >= 6) & (df["years_at_company"] <= 10),
    (df["years_at_company"] >= 11)
]


# 💻 Step 2 – Create Labels


tenure_choices = [
    "New Employee",
    "Established Employee",
    "Experienced Employee",
    "Veteran Employee"
]


# 💻 Step 3 – Create the Feature

df["employee_tenure_category"] = np.select(
    tenure_conditions,
    tenure_choices,
    default="Unknown"
)


# 💻 Step 4 – Verify

df[[
    "years_at_company",
    "employee_tenure_category"
]].head(10)




# 💻 Step 5 – Count

df["employee_tenure_category"].value_counts()

````


# ⭐ Feature 7 – Attrition Risk Score

This isn't machine learning yet. It's a **rule-based AI feature**, similar to what many companies build before training predictive models.

---

# 🏢 Business Problem

Imagine the HR Director asks:

> **"Can we identify employees who are at a high risk of leaving before they resign?"**

The original dataset doesn't have this answer.

So we'll create it.

---

# 🧠 Which columns should we use?

We should only use columns that have a logical relationship with attrition.

| Column              | Why it matters                                          |
| ------------------- | ------------------------------------------------------- |
| `overtime_status`   | Employees working overtime may be more likely to leave. |
| `job_satisfaction`  | Low satisfaction can increase attrition risk.           |
| `work_life_balance` | Poor balance often contributes to turnover.             |
| `promotion_due`     | Waiting too long for promotion can reduce engagement.   |

Notice I **didn't** include age or gender because there's no general business rule that those alone determine attrition risk.

---

# 🎯 Business Rules

### 🔴 High Risk

Employee is high risk if:

* Overtime = Yes
* Job Satisfaction ≤ 2
* Work-Life Balance ≤ 2
* Promotion Due = Yes

---

### 🟡 Medium Risk

Employee meets **any two** of the above conditions.

---

### 🟢 Low Risk

Everyone else.

---



---

# 🎤 Interview Answer

If the interviewer asks:

> **"How did you create the Attrition Risk Score?"**

You can answer:

> "I designed a rule-based attrition risk score by combining four HR factors: overtime status, job satisfaction, work-life balance, and promotion eligibility. Employees meeting all four risk conditions were classified as High Risk, those meeting at least two conditions as Medium Risk, and the remaining employees as Low Risk. This engineered feature was later used to generate AI-driven workforce insights."

---

# ⭐ Why this feature is special

This is no longer simple feature engineering.

You're building a **Decision Support System**.

```
Employee Data
       │
       ▼
Business Rules
       │
       ▼
Attrition Risk Score
       │
       ▼
AI Insights
       │
       ▼
Executive Dashboard


````python
# 💻 Step 1 – Conditions

risk_conditions = [
    (
        (df["over_time"] == "Yes") &
        (df["job_satisfaction"] <= 2) &
        (df["work_life_balance"] <= 2) &
        (df["promotion_due"] == "Yes")
    ),

    (
        (
            (df["over_time"] == "Yes").astype(int) +
            (df["job_satisfaction"] <= 2).astype(int) +
            (df["work_life_balance"] <= 2).astype(int) +
            (df["promotion_due"] == "Yes").astype(int)
        ) >= 2
    )
]


````

````python
# 💻 Step 2 – Labels


risk_choices = [
    "High Risk",
    "Medium Risk"
]


# 💻 Step 3 – Create the Feature


df["attrition_risk_score"] = np.select(
    risk_conditions,
    risk_choices,
    default="Low Risk"
)




# 💻 Step 4 – Verify

df[[
    "over_time",
    "job_satisfaction",
    "work_life_balance",
    "promotion_due",
    "attrition_risk_score"
]].head(15)


# 💻 Step 5 – Count

df["attrition_risk_score"].value_counts()
````
